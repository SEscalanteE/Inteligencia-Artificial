## Gaussian Process Regressor (GPR)

### Conceptos Clave:

- **Proceso Gaussiano (GP):** Generalización de la distribución normal multivariada a dimensiones infinitas, usado para modelar funciones con una función de media $m(\mathbf{x})$ y una función de covarianza $k(\mathbf{x}, \mathbf{x}')$.
  
### Inferencia y Predicción:

- **Distribución conjunta:** La distribución conjunta de las observaciones y la predicción sigue una distribución normal.
- **Predicción:** Se realiza condicionando sobre los datos observados para obtener la media y la varianza predictivas de los nuevos puntos.

### Ventajas y Desventajas:

- **Ventajas:** Predicciones probabilísticas, flexibilidad para modelar funciones complejas.
- **Desventajas:** Escalabilidad limitada por la inversión de matrices, la elección del kernel es crucial.

## Gaussian Process Classification (GPC)

### Modelo:

- En lugar de predecir valores continuos, en GPC se predicen etiquetas discretas utilizando una función latente $f(\mathbf{x})$ que sigue un proceso Gaussiano, con una función de enlace sigmoide para obtener probabilidades.

### Inferencia:

- **Distribución posterior:** La distribución posterior no es analíticamente tractable, por lo que se usan métodos de aproximación como Laplace Approximation, Expectation Propagation (EP) y Variational Inference.

### Predicción:

- La predicción se basa en calcular la probabilidad posterior de que una nueva entrada pertenezca a una clase.

### Ventajas y Desventajas:

- **Ventajas:** Predicciones probabilísticas, flexibilidad para modelar funciones no lineales.
- **Desventajas:** Inferencia más compleja, escalabilidad limitada, elección crítica del kernel.

## GP con TensorFlow: GPFlow

### Gaussian Process Regression (GPR) con GPFlow:

GPFlow facilita la implementación de GPR utilizando la optimización de máxima verosimilitud marginal (MLE). Un modelo básico se construye especificando un kernel y datos de entrenamiento.

### Implementación:

1. **Kernel:** Por ejemplo, el kernel RBF.
2. **Modelo GPR:** Creado con los datos de entrenamiento y el kernel.
3. **Optimización:** Se utiliza un optimizador de GPFlow para ajustar los parámetros del modelo.

### Predicción:

Una vez entrenado el modelo, se puede realizar predicciones obteniendo la media predictiva y la varianza.

### Ventajas y Desventajas:

- **Ventajas:** Predicciones probabilísticas, integración con TensorFlow, fácil de usar.
- **Desventajas:** Escalabilidad limitada, ajuste de hiperparámetros crucial.


#Gaussian Process Regressor

1 Pruebe otras combinaciones de kernels para entrenar el regresor.

Combinación usada:
```
          k(x, x') = k_const(1) * k_rbf(1)
         + k_const(1) * k_dotproduct(1)
         + k_const(1) * k_expsinesquared(1, 1)
         + k_const(1) * k_rationalquadratic(1, 0.5)
         + k_const(1) * k_matern(1, 1.5)
         + k_whitekernel(0.01)

```
Descripción de los kernels:

* k_const(1): Kernel constante con un valor constante de 1
* k_rbf(1): Kernel radial (RBF) con longitud de escala igual a 1
* k_dotproduct(1): Kernel de producto escalar con parámetro σ₀ igual a 1
* k_expsinesquared(1, 1): Kernel exponencial seno al cuadrado con longitud de escala 1 y periodicidad 1
* k _rationalquadratic(1, 0.5): Kernel racional cuadrático con longitud de escala 1 y α igual a 0.5
* k_matern(1, 1.5): Kernel Matern con longitud de escala 1 y parámetro ν igual a 1.5
* k_whitekernel(0.01): Kernel de ruido blanco con nivel de ruido 0.01



````python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel, DotProduct, ExpSineSquared, RationalQuadratic, Matern, WhiteKernel
from sklearn.model_selection import ShuffleSplit
import matplotlib

# Función para generar los datos
def generate_data(N=500, A=1, B=1, F1=1, F2=2, SNR_db=2):
    X = np.linspace(0, 1.2, N).reshape(-1, 1)
    y_clean = A * np.sin(2 * np.pi * F1 * X) + B * np.cos(2 * np.pi * F2 * X)
    SNR_linear = 10**(SNR_db / 10)
    signal_power = np.mean(y_clean**2)
    noise_power = signal_power / SNR_linear
    noise_std = np.sqrt(noise_power)
    y_noisy = y_clean + np.random.normal(0, noise_std, N).reshape(-1, 1)
    return X, y_noisy, y_clean

# Generar los datos
Xdata, ydata, y_clean = generate_data()

# Particionar los datos utilizando ShuffleSplit
rs = ShuffleSplit(n_splits=1, random_state=0, test_size=0.3)
for i, (train_i, test_i) in enumerate(rs.split(Xdata)):
    print(f"Iteración {i} - Entrenamiento y Prueba")

X_train, X_test, y_train, y_test = Xdata[train_i], Xdata[test_i], ydata[train_i], ydata[test_i]

# Mostrar las dimensiones de los datos de entrenamiento y prueba
print(f"Dimensión de X_train: {X_train.shape}")
print(f"Dimensión de X_test: {X_test.shape}")

# Definir el kernel
kernel = (ConstantKernel(constant_value=1.0, constant_value_bounds=(0.0, 10.0)) * RBF(length_scale=1, length_scale_bounds=(1e-1, 10.0)) +
          ConstantKernel(constant_value=1.0, constant_value_bounds=(0.0, 10.0)) * DotProduct(sigma_0=1.0, sigma_0_bounds=(1e-05, 100000.0)) +
          ConstantKernel(constant_value=1.0, constant_value_bounds=(0.0, 10.0)) * ExpSineSquared(length_scale=1.0, periodicity=1.0, length_scale_bounds=(1e-05, 100000.0), periodicity_bounds=(1e-05, 100000.0)) +
          ConstantKernel(constant_value=1.0, constant_value_bounds=(0.0, 10.0)) * RationalQuadratic(length_scale=1.0, alpha=0.5, length_scale_bounds=(1e-2, 10.0)) +
          ConstantKernel(constant_value=1.0, constant_value_bounds=(0.0, 10.0)) * Matern(length_scale=1.0, nu=1.5, length_scale_bounds=(1e-1, 10.0)) +
          WhiteKernel(noise_level=1e-2, noise_level_bounds=(1e-10, 1e1)))

# Definir el modelo GaussianProcessRegressor
gpr = GaussianProcessRegressor(kernel=kernel, alpha=0.1)

# Entrenar el modelo
gpr.fit(X_train, y_train)

# Predecir en los datos de prueba
X_test_sorted = np.sort(X_test, axis=0)
y_pred, y_std = gpr.predict(X_test_sorted, return_std=True)

# Graficar resultados de la predicción
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Datos de entrenamiento y prueba
axs[0].plot(X_train, y_train, 'r.', label='Entrenamiento')
axs[0].plot(X_test, y_test, 'b.', label='Prueba')
axs[0].legend()
axs[0].set_xlabel('x')
axs[0].set_ylabel('y=f(x)')
axs[0].set_title('Partición de datos')

# Subplot 2: Predicción del Gaussian Process
axs[1].plot(X_test_sorted, y_pred, label='Predicción GPR', color='purple')
axs[1].fill_between(X_test_sorted.ravel(), y_pred.ravel() - 1.96*y_std, y_pred.ravel() + 1.96*y_std, alpha=0.2, color='purple', label='Intervalo de confianza 95%')
axs[1].scatter(X_test, y_test, color='orange', label='Datos de prueba (ruidosos)')
axs[1].legend()
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Predicción del Gaussian Process')

plt.tight_layout()
plt.show()

# Imprimir el kernel ajustado y el log-marginal likelihood
print(f"Kernel ajustado después de entrenar: \n{gpr.kernel_}")
print(f"Log-marginal likelihood: {gpr.log_marginal_likelihood(gpr.kernel_.theta):.3f}")

# Analizar la influencia del hiperparámetro alpha
alphas = [1e-4, 1e-2, 1, 1e2, 1e4]

for alpha in alphas:
    print(f"\nAlpha: {alpha}")
    gpr = GaussianProcessRegressor(kernel=kernel, alpha=alpha, random_state=0)
    gpr.fit(X_train, y_train)
    y_pred, y_std = gpr.predict(X_test_sorted, return_std=True)

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Subplot 1: Datos de entrenamiento y prueba
    axs[0].plot(X_train, y_train, 'r.', label='Entrenamiento')
    axs[0].plot(X_test, y_test, 'b.', label='Prueba')
    axs[0].legend()
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('y=f(x)')
    axs[0].set_title('Partición de datos')

    # Subplot 2: Predicción del Gaussian Process con diferente alpha
    axs[1].plot(X_test_sorted, y_pred, label=f'Predicción GPR (alpha={alpha})', color='purple')
    axs[1].fill_between(X_test_sorted.ravel(), y_pred.ravel() - 1.96*y_std, y_pred.ravel() + 1.96*y_std, alpha=0.2, color='purple')
    axs[1].scatter(X_test, y_test, color='orange', label='Datos de prueba (ruidosos)')
    axs[1].legend()
    axs[1].set_xlabel('x')
    axs[1].set_ylabel('y')
    axs[1].set_title(f'Predicción con alpha={alpha}')

    plt.tight_layout()
    plt.show()
````

````python
# Importar las bibliotecas
import numpy as np
import tensorflow as tf
import gpflow
import matplotlib.pyplot as plt
import seaborn as sns
from gpflow.utilities import print_summary
from tensorflow.keras.datasets import fashion_mnist
from sklearn.metrics import classification_report, confusion_matrix
````

````python
# Función para cargar y preparar los datos de FashionMNIST
def preprocess_fashion_mnist():
    (train_data, train_labels), (test_data, test_labels) = fashion_mnist.load_data()

    # Normalizar y aplanar las imágenes de 28x28 a vectores de 784 dimensiones
    train_data = train_data.reshape(-1, 28 * 28).astype(np.float64) / 255.0
    test_data = test_data.reshape(-1, 28 * 28).astype(np.float64) / 255.0

    return (train_data, train_labels), (test_data, test_labels)
````

````python
# Cargar y preprocesar los datos
(train_data, train_labels), (test_data, test_labels) = preprocess_fashion_mnist()
````

````python
# Reducir el conjunto de datos para acelerar la implementación
sample_size = 3000
train_data_small = train_data[:sample_size]
train_labels_small = train_labels[:sample_size]
````

````python
# Asegurar que las etiquetas sean del tipo correcto
train_labels_small = train_labels_small[:, None].astype(np.int64)
test_labels = test_labels[:, None].astype(np.int64)
````

````python
# Mostrar formas de los datos procesados
print(f"Tamaño de train_data_small: {train_data_small.shape}")
print(f"Tamaño de train_labels_small: {train_labels_small.shape}")
````

````python
# Definir el kernel y el modelo de GPflow
kernel = gpflow.kernels.Matern12()
inducing_points = train_data_small[:500]  # Se utilizan 500 puntos de inducción
````

````python
# Configurar la función de likelihood para clasificación multiclase
likelihood = gpflow.likelihoods.Softmax(num_classes=10)
````

````python
# Construir el modelo SVGP
model = gpflow.models.SVGP(kernel=kernel,
                           likelihood=likelihood,
                           inducing_variable=inducing_points,
                           num_latent_gps=10)
````

````python
# Mostrar el resumen del modelo
print_summary(model)
````

````python
# Parámetros de entrenamiento
epochs = 200
learning_rate = 0.1
optimizer = tf.optimizers.Adam(learning_rate=learning_rate)
history_loss = []
````

````python
# Función para realizar un paso de optimización
@tf.function
def train_step(model, X_batch, Y_batch):
    with tf.GradientTape() as tape:
        elbo_loss = -model.elbo((X_batch, Y_batch))
    gradients = tape.gradient(elbo_loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return elbo_loss
````

````python
# Ciclo de entrenamiento
for epoch in range(epochs):
    current_loss = train_step(model, train_data_small, train_labels_small)
    history_loss.append(current_loss.numpy())

    if epoch % 10 == 0:
        print(f"Época {epoch}, Pérdida: {current_loss.numpy():.4f}")
````

````python
# Visualizar la historia de la pérdida
plt.plot(history_loss)
plt.title('Historia de la Pérdida durante el Entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Pérdida (ELBO)')
plt.show()
````

````python
# Evaluar el modelo en el conjunto de pruebas
test_data_tensor = tf.convert_to_tensor(test_data)
mean_predictions, _ = model.predict_y(test_data_tensor)
predicted_classes = np.argmax(mean_predictions, axis=1)
accuracy = np.mean(predicted_classes.reshape(-1, 1) == test_labels)
print(f'Precisión en el conjunto de pruebas: {accuracy * 100:.2f}%')
````

````python
# Generar la matriz de confusión y el reporte de clasificación
test_labels_flat = test_labels.flatten()
conf_matrix = confusion_matrix(test_labels_flat, predicted_classes)
````

````python
# Normalizar la matriz de confusión y mostrarla
conf_matrix_percent = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis] * 100

plt.figure(figsize=(7, 6))
sns.heatmap(conf_matrix_percent, annot=True, fmt='.2f', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
plt.title('Matriz de Confusión Normalizada (%)')
plt.xlabel('Predicciones')
plt.ylabel('Etiquetas Verdaderas')
plt.show()

# Reporte de clasificación
classification_rep = classification_report(test_labels_flat, predicted_classes, digits=4)
print("\nReporte de Clasificación:\n")
print(classification_rep)
````

````python
import tensorflow as tf

def mmd_loss(y_true, y_pred, sigma=1.0):
    """
    Calcula la pérdida MMD entre y_true y y_pred.

    y_true: etiquetas verdaderas (one-hot vectors)
    y_pred: etiquetas predichas (probabilidades softmax)
    sigma: desviación estándar del kernel Gaussiano
    """
    # Asegurarse de que y_true y y_pred son tensores de tipo float32
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    def cdist(XA, XB):
        """
        Calcula la distancia euclidiana entre cada par de puntos de XA y XB.
        """
        # Expandir las dimensiones para hacer las diferencias entre todos los pares
        XA_exp = tf.expand_dims(XA, axis=1)  # (m, 1, d)
        XB_exp = tf.expand_dims(XB, axis=0)  # (1, n, d)
        # Calcular la distancia euclidiana
        dist_matrix = tf.sqrt(tf.reduce_sum(tf.square(XA_exp - XB_exp), axis=-1))
        return dist_matrix
    # Definir el kernel Gaussiano
    def gaussian_kernel_matrix(x, y, sigma=1.0):
        kernel = tf.exp(-cdist(x,y) / (2 * sigma ** 2))
        return kernel
    # Calcular las matrices de kernel
    K_yy = gaussian_kernel_matrix(y_true, y_true)
    K_ff = gaussian_kernel_matrix(y_pred, y_pred)
    K_yf = gaussian_kernel_matrix(y_true, y_pred)
    # Calcular la MMD: usar reduce_mean para obtener un escalar
    mmd = tf.reduce_mean(K_yy + K_ff - 2 *K_yf)

    return mmd,K_yy

# Ejemplo de uso con tensores de prueba
y_true_example = tf.random.uniform((10, 5))
y_pred_example = tf.random.uniform((10, 5))
ind= np.argsort(y_test[0:500])
loss,k_yy = mmd_loss(tf.keras.utils.to_categorical(y_test[0:500][ind], num_classes=10), tf.keras.utils.to_categorical(y_test[0:500][ind], num_classes=10), sigma=1.0)
print(f"Pérdida MMD: {loss.numpy()}")
plt.imshow(k_yy.numpy(
))
plt.colorbar()
````

````python
!pip install -U git+https://github.com/UN-GCPDS/python-gcpds.EEG_Tensorflow_models

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Conv2D, Flatten, Dense, Lambda, Reshape, Conv2DTranspose, BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import layers, models, initializers, regularizers
from tensorflow.keras.losses import binary_crossentropy
from tensorflow.keras.datasets import mnist
from tensorflow.keras.optimizers import Adam
from keras.losses import MeanSquaredError
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
import matplotlib as mpl
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
from matplotlib import cm
from tf_keras_vis.gradcam_plus_plus import GradcamPlusPlus
from tf_keras_vis.utils.model_modifiers import ReplaceToLinear
from tf_keras_vis.utils.scores import CategoricalScore
from tensorflow.keras.losses import Loss
from tensorflow.keras.utils import to_categorical
import tensorflow.keras.backend as K

# Load the dataset
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
X_train_full = X_train_full.astype(np.float32) / 255
X_test = X_test.astype(np.float32) / 255
X_train, X_valid = X_train_full[:-5000], X_train_full[-5000:]
y_train, y_valid = y_train_full[:-5000], y_train_full[-5000:]

# Ensure correct shape (batch_size, 28, 28, 1)
X_train = np.expand_dims(X_train, -1)
X_valid = np.expand_dims(X_valid, -1)

# One-hot encoding for labels
num_classes = 10
y_train_one_hot = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_valid_one_hot = tf.keras.utils.to_categorical(y_valid, num_classes=num_classes)

# Plot sample images
plt.figure(figsize=(10, 5))
categories = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
              'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

for i in range(10):
    example_index = np.where(y_train == i)[0][0]
    image = X_train[example_index]

    plt.subplot(1, 10, i + 1)
    plt.imshow(image, cmap='gray')
    plt.title(categories[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

@tf.keras.utils.register_keras_serializable()
class Sampling(tf.keras.layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

def plot_mnist_autoencoder(x, x_, y=None, y_=None, cmap='gray', vmin=0, vmax=1):
    plt.figure(figsize=(20, 4))
    num_images = len(x)

    for i in range(num_images):
        plt.subplot(2, num_images, i + 1)
        plt.imshow(x[i].reshape(28, 28), cmap=cmap, vmin=vmin, vmax=vmax)
        if y is not None:
            plt.title(f'True: {y[i]}')
        plt.axis('off')

        plt.subplot(2, num_images, num_images + i + 1)
        plt.imshow(x_[i].reshape(28, 28), cmap=cmap, vmin=vmin, vmax=vmax)
        if y_ is not None:
            color = "red" if y is not None and y[i] != y_[i] else 'green'
            plt.title(f'Pred: {y_[i]}', color=color)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

def compare_images(original_images, predicted_images, true_labels, label_names, indices, figsize=(10, 5)):
    plt.figure(figsize=figsize)
    num_images = len(indices)

    for i, index in enumerate(indices):
        plt.subplot(2, num_images, i + 1)
        plt.imshow(original_images[index].squeeze(), cmap='gray')
        plt.title(f'Original {index}\nLabel: {label_names[true_labels[index]]}')
        plt.axis('off')

        plt.subplot(2, num_images, i + 1 + num_images)
        plt.imshow(predicted_images[index].squeeze(), cmap='gray')
        plt.title(f'Reconstrucción {index}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Loss functions
def reconstruction_loss(y_true, y_pred):
    y_true_flat = tf.keras.backend.flatten(y_true)
    y_pred_flat = tf.keras.backend.flatten(y_pred)
    return tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_true_flat, y_pred_flat))

def dkl_loss(y_true, y_pred):
    latent_dim = y_pred.shape[1] // 2
    z_mean = y_pred[:, :latent_dim]
    z_log_var = y_pred[:, latent_dim:]
    return -0.5 * tf.reduce_mean(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))

def dependency_criterion(images, predicted_labels):
    f_mean = K.mean(images, axis=0)
    g_mean = K.mean(predicted_labels, axis=0)
    f_centered = images - f_mean
    g_centered = predicted_labels - g_mean
    batch_size = tf.cast(tf.shape(images)[0], tf.float32)
    cov_xy = K.dot(K.transpose(f_centered), g_centered) / batch_size
    loss = -tf.reduce_sum(tf.linalg.diag_part(cov_xy))
    return loss

def model_class(input_shape=(28, 28, 1), num_classes=10, latent_dim=128):
    inputs = tf.keras.layers.Input(shape=input_shape)
    x = tf.keras.layers.Conv2D(6, kernel_size=3, activation='selu',
                               kernel_initializer=tf.keras.initializers.GlorotUniform(seed=12),
                               padding='same', name='conv1')(inputs)
    x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2))(x)
    x = tf.keras.layers.Dropout(0.15)(x)
    x = tf.keras.layers.Conv2D(16, kernel_size=3, activation='selu',
                               kernel_initializer=tf.keras.initializers.GlorotUniform(seed=23),
                               padding='same', name='conv2')(x)
    x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2))(x)
    x = tf.keras.layers.Conv2D(120, kernel_size=3, activation='selu',
                               kernel_initializer=tf.keras.initializers.GlorotUniform(seed=52),
                               padding='same', name='conv3')(x)
    x = tf.keras.layers.Flatten()(x)
    x = tf.keras.layers.Dropout(0.15)(x)

    z_mean = tf.keras.layers.Dense(latent_dim, name='z_mean')(x)
    z_log_var = tf.keras.layers.Dense(latent_dim, name='z_log_var')(x)
    z_gaussian = tf.keras.layers.Concatenate(name='z_gaussian', axis=-1)([z_mean, z_log_var])
    z = Sampling()([z_mean, z_log_var])
    softmax_output = tf.keras.layers.Dense(num_classes, activation='softmax', name='softmax_output')(z)

    x = tf.keras.layers.Dense(7 * 7 * 16, activation='selu')(z)
    x = tf.keras.layers.Reshape((7, 7, 16))(x)
    x = tf.keras.layers.Conv2DTranspose(16, kernel_size=3, activation='selu',
                                        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=12),
                                        padding='same')(x)
    x = tf.keras.layers.Dropout(0.15)(x)
    x = tf.keras.layers.UpSampling2D(size=(2, 2))(x)
    x = tf.keras.layers.Conv2DTranspose(6, kernel_size=3, activation='selu',
                                        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=23),
                                        padding='same', name='conv2Transpose')(x)
    x = tf.keras.layers.Dropout(0.15)(x)
    x = tf.keras.layers.UpSampling2D(size=(2, 2))(x)
    reconstruction = tf.keras.layers.Conv2D(input_shape[2], kernel_size=3, padding='same',
                                            activation='sigmoid', name='reconstruction')(x)

    vae = tf.keras.models.Model(inputs, [reconstruction, z_gaussian, softmax_output], name='vae')
    return vae

input_shape = (28, 28, 1)
num_classes = 10
latent_dim = 64
learning_rate = 1e-3

model = model_class(input_shape, num_classes, latent_dim)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              loss={'reconstruction': reconstruction_loss,
                    'z_gaussian': dkl_loss,
                    'softmax_output': dependency_criterion},
              loss_weights={'reconstruction': 0.5,
                            'z_gaussian': 0.1,
                            'softmax_output': 0.4},
              metrics={'reconstruction': [reconstruction_loss],
                       'z_gaussian': [dkl_loss],
                       'softmax_output': [dependency_criterion]})

seed = 42
epochs = 50
batch_size = 32
learning_rate = 1e-2

history = model.fit(X_train,
                    [X_train, X_train, y_train_one_hot],
                    epochs=epochs,
                    batch_size=batch_size,
                    validation_data=(X_valid, [X_valid, X_valid, y_valid_one_hot]))

loss = history.history['loss']
val_loss = history.history['val_loss']
reconstruction_loss = history.history['reconstruction_reconstruction_loss']
val_reconstruction_loss = history.history['val_reconstruction_reconstruction_loss']
classification_loss = history.history['softmax_output_dependency_criterion']
val_classification_loss = history.history['val_softmax_output_dependency_criterion']
dkl_loss = history.history['z_gaussian_dkl_loss']
val_dkl_loss = history.history['val_z_gaussian_dkl_loss']

plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 3)
plt.plot(classification_loss, label='Training Classification Loss')
plt.plot(val_classification_loss, label='Validation Classification Loss')
plt.title('Classification Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

X_re = model.predict(X_test)

softmax_output = X_re[2]
predicted_labels = np.argmax(softmax_output, axis=1)

accuracy = accuracy_score(y_test, predicted_labels)
print("Accuracy:", accuracy)
print(classification_report(y_test, predicted_labels))
````
